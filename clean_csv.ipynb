```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
#Read in the drivers csv file
drivers_df = pd.read_csv("Resources_Original/drivers.csv")
drivers_df.head()
```

```python
#Rename forename to first_name and surname to last_name
drivers_df.rename(columns={'forename':'first_name'}, inplace=True)
drivers_df.rename(columns={'surname':'last_name'}, inplace=True)
drivers_df.head()
```

```python
#Eliminate the "url" column and the "number" column
drivers_df_new = drivers_df.drop(columns=['url'])
drivers_df_new = drivers_df_new.drop(columns=['number'])
drivers_df_new.head()
```

```python
#Drop the dob column
#Drop null values
drivers_df_clean = drivers_df_new.drop(columns=['dob'])
drivers_df_clean = drivers_df_clean.dropna()
drivers_df_clean
```

```python
#Reset the index after dropping rows
#Eliminate special characters from the column names by using the replace() method
drivers_df_clean.reset_index(drop=True, inplace=True)
drivers_df_clean.columns = drivers_df_clean.columns.str.replace('[#,@,&,+,*,%,$,^,!,?,>,<,=,_,:,;,(,),]', '')
drivers_df_clean.head()
```

```python
#Remove code column
drivers_df_clean = drivers_df_clean.drop(columns=['code'])
drivers_df_clean.head()
```

```python
#Export to CSV
drivers_df_clean.to_csv('Cleaned_Resources/drivers.csv', index=False)
```

## Starting SQLite

```python
# Connect to the SQLite database
connection = sqlite3.connect('sqlite_database/f1_database.sqlite')
cursor = connection.cursor()
```

```python
# Create Tables
cursor.execute('''
    CREATE TABLE IF NOT EXISTS Drivers (
        driverId INTEGER PRIMARY KEY,
        driverRef TEXT,
        firstname TEXT,
        lastname TEXT,
        nationality TEXT
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Teams (
        teamId INTEGER PRIMARY KEY,
        name TEXT,
        nationality TEXT
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Circuits (
        circuitId INTEGER PRIMARY KEY,
        name TEXT,
        location TEXT,
        country TEXT,
        lat REAL,
        lng REAL
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS Races (
        raceId INTEGER PRIMARY KEY,
        year INTEGER,
        round INTEGER,
        circuitId INTEGER,
        name TEXT,
        FOREIGN KEY (circuitId) REFERENCES Circuits(circuitId)
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS SprintResults (
        resultId INTEGER PRIMARY KEY,
        raceId INTEGER,
        driverId INTEGER,
        teamId INTEGER,
        grid INTEGER,
        position TEXT,
        points INTEGER,
        laps INTEGER,
        time TEXT,
        fastestLap TEXT,
        fastestLapTime TEXT,
        FOREIGN KEY (raceId) REFERENCES Races(raceId),
        FOREIGN KEY (teamId) REFERENCES Teams(teamId)
    )
''')

cursor.execute('''
        CREATE TABLE IF NOT EXISTS results (
        resultId INTEGER PRIMARY KEY,
        raceId INTEGER,
        driverId INTEGER,
        teamId INTEGER,
        grid INTEGER,
        position TEXT,
        points INTEGER,
        laps INTEGER,
        time TEXT,
        fastestLap TEXT,
        rank TEXT,
        fastestLapTime TEXT,
        fastestLapSpeed TEXT,
        FOREIGN KEY (raceId) REFERENCES Races(raceId),
        FOREIGN KEY (teamId) REFERENCES Teams(teamId)
    )
''' )

cursor.execute('''
    CREATE TABLE IF NOT EXISTS driver_standings (
        driverStandingsId INTEGER PRIMARY KEY,
        raceId INTEGER,
        driverId INTEGER,
        points TEXT,
        position TEXT,
        FOREIGN KEY (raceId) REFERENCES Races(raceId)
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS team_standings (
        teamStandingsId INTEGER PRIMARY KEY,
        raceId INTEGER,
        teamId INTEGER,
        points TEXT,
        position TEXT,
        FOREIGN KEY (raceId) REFERENCES Races(raceId),
        FOREIGN KEY (teamId) REFERENCES Teams(teamId)
    )
''')
```
